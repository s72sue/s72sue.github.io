```python
# Leaky Integrate and Fire Neuron Model
# Random unit vectors - around a unit circle - 100 NEURONS

%pylab inline
import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from scipy.optimize import curve_fit

class LeakyIntegNeuron ():
     
    def __init__(self, tau_ref=0.002, tau_rc=0.02):
        self.tau_rc = tau_rc
        self.tau_ref = tau_ref
    
    def rates (self, x, gain, bias, S):
        J = (gain * x) + bias       # compute the current using the gain and the bias
        a = np.linspace(0,0,S)
        try:  
            i = 0 
            for element in J:
                if element > 1:
                    a[i] = 1.0 / (self.tau_ref - self.tau_rc * math.log( 1 - (1/element) ) )
                i = i+1    
        finally:
            np.seterr(divide='ignore', invalid='ignore')
        return a

    def cal_bias(self, intercept, gain):
        bias = 1 - (gain*intercept)  
        return bias

    def cal_gain(self, intercept, max_firing_rate):
        encoder = random.choice([-1,1])           # encoder e indicating preferred direction 
        try:
            x = 1.0 / (1 - np.exp((self.tau_ref - (1.0 / max_firing_rate)) / self.tau_rc))
            gain = ( (1-x) / (intercept-1.0) )
        finally:
            np.seterr(divide='ignore')
        return gain
 

def cal_decoders(activity_matrix, x, S, std_deviation):
    gamma = (( np.dot(activity_matrix, activity_matrix.transpose()) ) / S )            # Y = AA^T/S
   
    # if standard deviation is not zero, then consider the effect of noise
    if std_deviation:  
        gamma = gamma + (std_deviation**2)
           
    upsilon = ( np.dot(activity_matrix, x) ) / S         # v = A.x/S
    try:
        decoders = np.dot(np.linalg.pinv(gamma), upsilon)        # compute the decoders (using psedo inverse to deal with singular matrices)
    finally:
        np.seterr(divide='ignore', invalid='ignore')
    return decoders 


# computes the xhat vector 
def compute_xhat(x, decoders, activity_matrix, S):
    xhat = np.empty([S,2])
    i = 0
    for count in xrange(S):
        xhat[i,:] = np.dot(activity_matrix[:,i].transpose(), decoders)
        i = i+1  
    return xhat


def get_rand_xvalues():
    theta = 2*np.pi*np.random.random()
    u = np.random.random() + np.random.random()
    if (u>1):
        r = 2 - u
    else:
        r = u
    return [r*cos(theta), r*sin(theta)]

def get_random_vec(dims):
    v = np.random.standard_normal(dims)
    return v / np.linalg.norm(v)


def main():
    
    neurons_num = 100
    S = 40              # number of samples = 100 in this case    
    intercepts = -1 + 2 * np.random.rand(neurons_num)       # intercepts of 16 neurons will be used to calculate the bias
    max_firing_rates = 100 + 100 * np.random.rand(neurons_num)     # max firing rates of neurons should be between 100 to 200
    neurons = [LeakyIntegNeuron() for count in xrange(neurons_num)]       # create 16 rectified linear neurons
    activity_matrix = np.zeros((neurons_num,S*S))
    
    
    encoders = np.array([get_random_vec(2) for count in xrange(neurons_num)])
    plt.figure()
    plt.axis('equal')
    plt.title("Encoders")
    for e in encoders:
        plt.plot([0,e[0]], [0,e[1]], 'r')
    
    a = np.linspace(-1,1,S)  
    x = np.zeros((S*S,2))
    
    
    # creating tiled 2D x values
    m = 0
    n = 0
    for count in xrange(S):
        k = 0
        for count in xrange(S):
            x[n,0] = a[m]
            x[n,1] = a[k]
            n = n + 1
            k = k + 1
        m = m + 1
  
    
    i = 0
    for neuron in neurons:       
        gain = neuron.cal_gain(intercepts[i], max_firing_rates[i])
        bias = neuron.cal_bias(intercepts[i], gain)
        firing_rates = neuron.rates(np.dot(x, encoders[i].transpose()), gain, bias, S*S)     # firing rates of a neuron over the entire range (range of x values)          
        activity_matrix[i,:] = firing_rates        # populate the firing rates of this neuron at every value of x  
            
        i = i + 1    
    
    # compute the decoder 
    std_deviation = 0.2*numpy.max(activity_matrix)  
    activity_matrix_noisy = activity_matrix + numpy.random.normal(loc=0, scale=std_deviation, size=activity_matrix.shape)
    decoders = cal_decoders(activity_matrix_noisy, x, S*S, std_deviation)    
    #print "\n Decoder Values are: \n",  decoders
    
    plt.figure()
    plt.axis('equal')
    plt.title("Decoders")
    for d in decoders:
        plt.plot([0,d[0]], [0,d[1]], 'r')
     
   
    # Ques 2.2c
    # computing the 20 x values over a unit circle
    S2 = 20
    activity_matrix2 = np.zeros((neurons_num,S2))
    plt.figure()
    plt.axis('equal')
    plt.title("x values over a unit circle")
    X = np.array([get_rand_xvalues() for count in xrange(S2)])   # 20 random vectors inside a unit circle
    for x in X:
        plt.plot([0,x[0]], [0,x[1]], 'r')
          
    i = 0
    for neuron in neurons:       
        gain = neuron.cal_gain(intercepts[i], max_firing_rates[i])
        bias = neuron.cal_bias(intercepts[i], gain)
        firing_rates = neuron.rates(np.dot(X, encoders[i].transpose()), gain, bias, S2)     # firing rates of a neuron over the entire range (range of x values)          
        activity_matrix2[i,:] = firing_rates        # populate the firing rates of this neuron at every value of x      
        i = i + 1 
   
    # compute xhat using decoders
    xhat = compute_xhat(X, decoders, activity_matrix2, S2)  
    plt.figure()
    #plt.axis('equal')
    plt.title("x (red) and xhat (green) values - using decoders")
    for x in X:
        plt.plot([0,x[0]], [0,x[1]], 'r')
        
    for x in xhat:
        plt.plot([0,x[0]], [0,x[1]], 'g') 
        
    rmse = np.sqrt(np.average((X-xhat)**2)) 
    print " RMSE (using decoders)\n" , rmse
    
    
    # Part d)
    # computing the 20 x values over a unit circle
    S3 = 20
    activity_matrix3 = np.zeros((neurons_num,S3))
    plt.figure()
    #plt.axis('equal')
    plt.title("x values over a unit circle")
    X = np.array([get_rand_xvalues() for count in xrange(S3)])   # vectors inside a unit circle
    for x in X:
        plt.plot([0,x[0]], [0,x[1]], 'b')
        
    i = 0
    for neuron in neurons:       
        gain = neuron.cal_gain(intercepts[i], max_firing_rates[i])
        bias = neuron.cal_bias(intercepts[i], gain)
        firing_rates = neuron.rates(np.dot(X, encoders[i].transpose()), gain, bias, S3)     # firing rates of a neuron over the entire range (range of x values)          
        activity_matrix3[i,:] = firing_rates        # populate the firing rates of this neuron at every value of x      
        i = i + 1 
   
    # compute xhat using encoders as decoders
    xhat2 = compute_xhat(X, encoders, activity_matrix3, S3)  
    plt.figure()
    #plt.axis('equal')
    plt.title("x values over a unit circle")
    for x in X:
        plt.plot([0,x[0]], [0,x[1]], 'r')
    
    figure()
    plt.title("xhat values using encoders as decoders")
    for x in xhat2:
        plt.plot([0,x[0]], [0,x[1]], 'g') 
        
    figure()
    for x in X:
        plt.plot([0,x[0]], [0,x[1]], 'r')
    
    plt.title("x (red) and xhat (green) values using encoders as decoders")
    for x in xhat2:
        plt.plot([0,x[0]], [0,x[1]], 'g') 
    
    rmse = np.sqrt(np.average((X-xhat2)**2)) 
    print " RMSE (using encoders as decoders) \n" , rmse
    
    
    # Part d.2 - Ignoring the magnitudes 
    
    S3 = 20
    activity_matrix4 = np.zeros((neurons_num,S3))
    plt.figure()
    plt.axis('equal')
    plt.title("x values over a unit circle - ignore magnitude")
    X = np.array([get_rand_xvalues() for count in xrange(S3)])   # vectors inside a unit circle
    for x in X:
        plt.plot([0,x[0]], [0,x[1]], 'b')
    #X = X / np.linalg.norm(X)
    
    i = 0
    for neuron in neurons:       
        gain = neuron.cal_gain(intercepts[i], max_firing_rates[i])
        bias = neuron.cal_bias(intercepts[i], gain)
        firing_rates = neuron.rates(np.dot(X, decoders[i].transpose()), gain, bias, S3)     # firing rates of a neuron over the entire range (range of x values)          
        activity_matrix4[i,:] = firing_rates        # populate the firing rates of this neuron at every value of x      
        i = i + 1 
   
    
    # compute xhat 
    X2 = X / np.linalg.norm(X)
    xhat3 = compute_xhat(X, decoders, activity_matrix4, S3)  
    xhat3 = xhat3 / np.linalg.norm(xhat3)
    plt.figure()
    #plt.axis('equal')
    plt.title("x (blue) and xhat (green) values - ignore magnitude decoder")  
    for x in X2:
        plt.plot([0,x[0]], [0,x[1]], 'b')
        
    for x in xhat3:
        plt.plot([0,x[0]], [0,x[1]], 'g') 
        
    rmse = np.sqrt(np.average((X2-xhat3)**2)) 
    print " RMSE (using decoders ignoring magnitudes) \n" , rmse    
    
    # ----------------------------------------------------------------------------------------------------------------
    plt.figure()
    xhat4 = xhat2 / np.linalg.norm(xhat2)
    plt.title("x (blue) and xhat (green) values - ignore magnitude encoder")
    X2 = X / np.linalg.norm(X)
    for x in X2:
        plt.plot([0,x[0]], [0,x[1]], 'b')
    
    for x in xhat4:
        plt.plot([0,x[0]], [0,x[1]], 'g')    
        
    rmse = np.sqrt(np.average((X2-xhat4)**2)) 
    print " RMSE (using encoders as decoders - ignore magnitude) \n" , rmse
    
if __name__ == "__main__": main()
```
