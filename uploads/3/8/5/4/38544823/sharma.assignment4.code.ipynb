# 1. Building an ensemble of neurons

```python
%pylab inline 

import numpy
import nengo
from nengo.utils.ensemble import tuning_curves
from nengo.dists import Uniform

def lif_neurons(tau_rc = 0.02, tau_ref = 0.002, radius = 1.0, flag = 0):

    N = 100
    lif_model = nengo.LIF(tau_rc=tau_rc, tau_ref=tau_ref)

    model = nengo.Network(label='Neurons')
    with model:
        neurons = nengo.Ensemble(N, dimensions=1, radius = radius, intercepts = Uniform(-1.0, 1.0), 
                                 max_rates = Uniform(100,200), neuron_type = lif_model) 
    
    sim = nengo.Simulator(model)
    x, A = tuning_curves(neurons, sim)

    if flag:
        figure()
        plot(x, A)
        xlabel('x')
        ylabel('firing rate (Hz)')

    with model:
        connection = nengo.Connection(neurons, neurons) #This is just to generate the decoders

    sim = nengo.Simulator(model)
    d = sim.data[connection].decoders.T
    x, A = tuning_curves(neurons, sim)
    xhat = numpy.dot(A, d)
    
    rmse = np.sqrt(np.average((x-xhat)**2))


    if flag:
        print 'RMSE with %d neurons is %g'%(N, rmse)
        figure()
        plot(x, x, 'b' , label = 'x')
        plot(x, xhat, 'g' , label = 'xhat')
        xlabel('$x$')
        ylabel('$\hat{x}$')
        legend()
        figure()
        plot(x, x-xhat, 'g')
        xlabel('$x$')
        ylabel('$x$ - $\hat{x}$')


    return rmse
    
def main():
    
    # Par a) - Tuning curves and representational accuracy plot
    lif_neurons(tau_rc = 0.02, tau_ref = 0.002, radius = 1.0, flag = 1)
    
    # ---------------------------------------------------------------------------------------------------------------------------
    # Part b) - Effect of changing the radius of neurons on RMSE
    
    rmse_array = np.zeros(4)
    radius = 1
    i = 0
    for value in range(4):
        rmse_array[i] = lif_neurons(tau_rc = 0.02, tau_ref = 0.002, radius = radius, flag = 0)
        radius = radius + 1
        i = i + 1
        
    figure()
    plot([1,2,3,4], rmse_array)
    xlabel("neuron radius")
    ylabel("RMSE")
    title ("RMSE with an increase in radius")
    #print rmse_array
        
    # ---------------------------------------------------------------------------------------------------------------------------
    # Part c) - Effect of change in tau_ref on RMSE
    
    rmse_array = np.zeros(7)
    tau_ref_array = np.zeros(7)
    tau_ref = 0.002
    i = 0
    for value in range(7):
        rmse_array[i] = lif_neurons(tau_rc = 0.02, tau_ref = tau_ref, radius = 1, flag = 0)
        tau_ref_array[i] = tau_ref
        #print tau_ref
        tau_ref = tau_ref + 0.0005
        i = i + 1
        
    figure()
    plot(tau_ref_array, rmse_array)
    xlabel("tau_ref")
    ylabel("RMSE")
    title ("RMSE with an increase in tau_ref")
     
    # ---------------------------------------------------------------------------------------------------------------------------
    # Part c) - Effect of change in tau_RC on RMSE
    
    rmse_array = np.zeros(7)
    tau_rc_array = np.zeros(7)
    tau_rc = 0.02
    i = 0
    for value in range(7):
        rmse_array[i] = lif_neurons(tau_rc = tau_rc, tau_ref = 0.002, radius = 1, flag = 0)
        tau_rc_array[i] = tau_rc
        #print tau_rc
        tau_rc = tau_rc + 0.02
        i = i + 1
        
    figure()
    plot(tau_rc_array, rmse_array)
    xlabel("tau_rc")
    ylabel("RMSE")
    title ("RMSE with an increase in tau_rc")
        
if __name__ == "__main__": main()
```

# 2. Connecting neurons

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from nengo.dists import Uniform
from nengo.utils.functions import piecewise
import nengo

def connnecting_neurons(stim_signal, flag=0):
    
    model = nengo.Network()

    with model:
        
        # create an input
        stim = nengo.Node(piecewise({0:stim_signal, 0.1:0 }))

        # Create two neuronal ensembles
        ens_A = nengo.Ensemble(100, dimensions=1, radius = 1, intercepts = Uniform(-1.0, 1.0), 
                               max_rates = Uniform(100,200), neuron_type = nengo.LIF(tau_rc = 0.02,tau_ref = 0.002))
        ens_B = nengo.Ensemble(50, dimensions=1, radius = 1, intercepts = Uniform(-1.0, 1.0), 
                               max_rates = Uniform(100,200), neuron_type = nengo.LIF(tau_rc = 0.02,tau_ref = 0.002))

        nengo.Connection(stim, ens_A, synapse = 0.01)       # Connect the input to the first neuronal ensemble
        
         # Connect the first neuronal ensemble to the second 
        if flag == 1:
            nengo.Connection(ens_A, ens_B, synapse = 0.01, function = lambda x: 1.0 - 2.0*x)    
        else:
            nengo.Connection(ens_A, ens_B, synapse = 0.01)     

    # get data from the simulation
    with model:
        stim_probe = nengo.Probe(stim)
        A_probe = nengo.Probe(ens_A, synapse=0.01)  
        B_probe = nengo.Probe(ens_B, synapse=0.01)


    sim = nengo.Simulator(model)
    sim.run(0.5)

    figure()
    plt.title("Input Signal")
    plt.plot(sim.trange(), sim.data[stim_probe])
    ylim(-0.2, 1.2)

    figure()
    plt.title("Ensemble A")
    plt.plot(sim.trange(), sim.data[A_probe])
    #ylim(-0.2, 1.2)

    figure()
    plt.title("Ensemble B")
    plt.plot(sim.trange(), sim.data[B_probe])
    #ylim(-0.2, 1.2)
    

def main():
    
    # Part a)
    stim_signal = 1.0
    connnecting_neurons(stim_signal, flag = 0)
    
    #Part b)
    stim_signal = 1.0
    connnecting_neurons(stim_signal, flag = 1)
    

if __name__ == "__main__": main()
```

## f)  Non Linear Dynamical Systems

### Rossler Attractor

A Rossler Attractor can be expressed by following differential equations:

$$
x'(t) = -(y(t) + z(t)) \\\
y'(t) = x(t) + a y(t) \\\
z'(t) = b + x(t)z(t) - c z(t)
$$

These equations can be written as follows where: 
$$\ x => x_0,  y => x_1,  z => x_2 $$

$$
\dot{x}_0 = -(x_1 + x_2) \\\
\dot{x}_1 = x_0 + a * x_1  \\\
\dot{x}_2 = b + x_0 x_2 - c * x_2  = b - x_2 (c - x_0)
$$

```python
import nengo
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

model = nengo.Network('Rossler Attractor', seed=3)

tau = 0.1
a = 0.5
b = 1
c = 13


"""a = 0.5
b = 0.5
c = 5.7"""

def feedback(x):
    
    # original
    dx0 = -x[1] - x[2]
    dx1 = x[0] + a*x[1]
    dx2 = b + x[2]*x[0] - c*x[2]
    return [dx0 * tau + x[0], 
            dx1 * tau + x[1], 
            dx2 * tau + x[2]]

with model:
    rossler = nengo.Ensemble(2000, dimensions=3, radius=60)
    
    nengo.Connection(rossler, rossler, function=feedback, synapse=tau)
    
    rossler_p = nengo.Probe(rossler, synapse=tau)
    
sim = nengo.Simulator(model)
sim.run(20)

figure(figsize=(12,4))
subplot(1,2,1)
plot(sim.trange(), sim.data[rossler_p]);
xlabel('Time (s)')
ylabel('State value')
       
subplot(1,2,2)
plot(sim.data[rossler_p][:,0], sim.data[rossler_p][:,1])
xlabel('$x_0$')
ylabel('$x_1$')

"""figure()
plot(sim.trange(), sim.data[rossler_p][:,0]);
figure()
plot(sim.trange(), sim.data[rossler_p][:,1]);
figure()
plot(sim.trange(), sim.data[rossler_p][:,2]);"""

"""fig = plt.figure()
ax = Axes3D(fig)
ax.plot(sim.data[rossler_p][:,2], 
        sim.data[rossler_p][:,0], 
        sim.data[rossler_p][:,1])"""

ax = plt.figure().add_subplot(111, projection='3d')
ax.plot(sim.data[rossler_p][:, 0],
        sim.data[rossler_p][:, 1],
        sim.data[rossler_p][:, 2])

ax.set_xlabel('$x_0$')
ax.set_ylabel('$x_1$')
ax.set_zlabel('$x_2$')
```

### Double Integrator

```python
%pylab inline
import matplotlib.pyplot as plt
from nengo.utils.functions import piecewise
from nengo.dists import Uniform
import nengo

model = nengo.Network(label='Integrator')
with model:
    
    stim = nengo.Node(piecewise({0: 0, 0.2: -1, 1: 0, 2: 2, 3: 0, 4: -2, 5: 0, 6: 1, 7: 0}))
    #stim = nengo.Node(piecewise({0: 0, 1: 1, 2: -2, 3: 2, 4: -2, 5: 2, 6: -2, 7: 1, 8: 0}))
    
    # An ensemble consisting of 200 neurons
    ens_A = nengo.Ensemble(200, dimensions=1, radius = 1, intercepts = Uniform(-1.0, 1.0), 
                           max_rates = Uniform(100,200), neuron_type = nengo.LIF())
    
    tau = 0.05
    # Connection of the input to the ensemble
    nengo.Connection(stim, ens_A, transform=tau, synapse=0.005) 

    # Connect the ensemble back to itself
    nengo.Connection(ens_A, ens_A, transform=1, synapse=tau) 
    
    # create another ensemble for creating a double integrator
    ens_B = nengo.Ensemble(200, dimensions=1, radius = 1, intercepts = Uniform(-1.0, 1.0), 
                           max_rates = Uniform(100,200), neuron_type = nengo.LIF())
    
    # Connection of the input to the ensemble
    nengo.Connection(ens_A, ens_B, transform=tau, synapse=0.005) 

    # Connect the ensemble back to itself
    nengo.Connection(ens_B, ens_B, transform=1, synapse=tau) 
    
    # Add probes
    input_probe = nengo.Probe(stim)
    A_probe = nengo.Probe(ens_A, synapse=0.01)
    B_probe = nengo.Probe(ens_B, synapse=0.01)
    
    
# Create the simulator and run the simulation for 1.5 seconds
sim = nengo.Simulator(model)
sim.run(9)    

plt.plot(sim.trange(), sim.data[input_probe], 'b', label="Input")
plt.plot(sim.trange(), sim.data[A_probe], 'r', label="Ensemble A output")
plt.plot(sim.trange(), sim.data[B_probe], 'k', label="Ensemble B output")
plt.legend(loc="best")
legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
ylim(-2.5, 2.5)
```
