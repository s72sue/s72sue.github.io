```python
%pylab inline
import random
import numpy as np
import functools
#import matplotlib.pyplot as plt
 

class GaussianWhiteNoise():
    
    def __init__(self, T, limit, rms):
        """
        Parameters:
        -----------
        T: the length of the signal in seconds
        limit: the maximum frequency for the signal (in Hz)
        rms: the root mean square power level of the signal.
        """
        self.T = T
        self.limit = limit
        self.rms = rms
              
    
    def create_sample(self, dt, rng):
        """
        Generate random fourier coefficients and convert them back to 
        time domain to get a signal with a limit on its max frequecy
        Parameters:
        -----------
        dt: the time step in seconds
        rng: A random number generatorn
        """
                    
        # set value for the std. deviation
        std_dev = 1
        
        # num of points(coefficients) is halved since irfft method used later returns n output points if input has n//2+1 points
        num_points = int(np.ceil(( self.T / dt) / 2.0 ))
      
        # generate the matrix containing the points
        y_values = rng.normal(0.0, std_dev, (1, num_points + 1) )
        x_values = rng.normal(0.0, std_dev, (1, num_points + 1) )
        points = x_values + 1j * y_values
        points[:, 0] = 0.0                  # set the DC component to zero 
       
             
        # Check that the max frequency is below the limit and adjust the power accordingly
        sample_freqs = np.fft.rfftfreq(num_points * 2, dt)      # compute discrete fourier transform sample frequencies  
                                                                # returns 501 frequency samples (n//2 + 1)       
        zero_points = sample_freqs > self.limit
        points[:, zero_points] = 0.0                            # set the coefficients to zero for frequencies higher than the limit
 
        a = np.array([0])
        signal = np.fft.irfft(points, axis=1)          # compute the inverse of the Discrete Fourier transform to get the real signal
        
        # compute the rms of the signal
        rms_orig = matplotlib.mlab.rms_flat(signal)
        
        # scale the signal so that it has the rms specified
        signal = signal * self.rms / rms_orig
          
        return functools.partial(self.sample_func, a, signal)


    @staticmethod
    def sample_func(a, signal):
        size =  signal.shape[1] - 1
        b = a % size
        sample = signal[:,np.array(b)[0]]
        a += 1
        return sample
        


# A function to generate a randomly varying x(t) and X(w) signal
def generate_signal(T, dt, rms, limit, seed):
    
    obj = GaussianWhiteNoise(T, limit, rms)
    sample = obj.create_sample(dt, rng=np.random.RandomState(seed))

    samples_array = np.arange(T/dt)             # generate evenly spaced samples
    x = np.array([sample() for i in samples_array])[:,0]
    
    X = np.fft.fftshift(np.fft.fft(x))          # compute the discrete fourier transform and 
                                                      # shift the zero frequency component to the center of the spectrum
    # plot the time domain signal
    figure()
    plot(samples_array*dt, x)          # samples_array is multiplied by dt to limit the duration of the signal to 'T'
    xlabel('$t$')
    ylabel('$x(t)$')
    
    return x, X
    
   

# Same as generate_signal but without plots
def gen_sig(T, dt, rms, limit, seed):
    obj = GaussianWhiteNoise(T, limit, rms)
    sample = obj.create_sample(dt, rng=np.random.RandomState(seed))

    samples_array = np.arange(T/dt)             # generate evenly spaced samples
    x = np.array([sample() for i in samples_array])[:,0]
    
    X = np.fft.fftshift(np.fft.fft(x))   # Fourier Transform in switch to frequency domain
    
    return x, X
 

# calculate and plot the average X omega value for 100 signals   
def plot_avgXomega(T, dt, rms, limit): 
    seed = 3
    sum_X = 0
    count = 100             # number of signals to compute the average
    for num in xrange(count):
        seed = seed + 1
        x, X = gen_sig(T, dt, rms, limit, seed)
        sum_X += np.abs(X)
    
    avg_X = sum_X / count
    freq = np.arange(int(T/dt)) - (T/dt)/2      # spread the frequencies evenly from -500 to 500 (enforcing symmetry to get a real signal)
    omega = 2*numpy.pi*freq


    figure()
    plot(omega, avg_X)
    xlabel('$\omega$')
    ylabel('$|X(\omega)|$')
    #xlim(-100,100)

    
def main():
    
    # Generating plots of x(t)
    T = 1
    dt = 0.001
    seed = 3
    rms = 0.5
    generate_signal(T, dt, rms, 5, seed)   
    generate_signal(T, dt, rms, 10, seed)
    generate_signal(T, dt, rms, 20, seed)
    
    
    # Generate plot of average X(omega)
    plot_avgXomega(T, dt, rms, 10)
    
   
            
if __name__ == "__main__": main()
```

```python
# Spiky Leaky Integrate and Fire Neuron Model

%pylab inline
import random
import numpy as np
import matplotlib.pyplot as plt

    
# subclass of LeakyIntegNeuron class    
class spikyLIF():
    
    def __init__(self, tau_rc=0.02, tau_ref=0.002):
        self.tau_rc = tau_rc
        self.tau_ref = tau_ref 
        
    def cal_bias(self):
        try:
            bias = 1.0 / (1 - np.exp((self.tau_ref - (1.0 / 40.0)) / self.tau_rc))
        finally:
            np.seterr(divide='ignore')    
        return bias

    def cal_gain(self, bias): 
        try:
            x = 1.0 / (1 - np.exp((self.tau_ref - (1.0 / 150.0)) / self.tau_rc))
            gain = x - bias
        finally:
            np.seterr(divide='ignore')
        return gain
    
    def cal_current(self, x, gain, encoder, bias):
        J = (gain * x * encoder) + bias  
        return J
    
    def generate_spikes(self, T, dt, J, Vth=1, V_spike=0.5, R=1):
        """
        Parameters:
        ----------
        T - total simulation time (sec)    
        dt - simulation time step (sec)  
        Vth - spike threshold voltage (V)
        V_spike - spike delta voltage above Vth (V)
        R  - resistance (Ohm)
        """
        time = np.arange(0, T, dt)       # time array
        t_rest = 0                      # initial refractory time
        
        V = zeros(len(time))    # potential (V) trace over time
        spike_time = list()
        
        num_spikes = 0
        test = 0
        # iterate over each time step
        for i, t in enumerate(time):  
            if t > t_rest:
                V[i] = V[i-1] + ((J*R - V[i-1]) / self.tau_rc) * dt
                test = test + 1
            if V[i] >= Vth:
                num_spikes += 1
                V[i] += V_spike
                t_rest = t + self.tau_ref
                spike_time.append(t)
            if V[i] < 0:  
                V[i] = 0

        print "Number of Spikes = " , num_spikes  #, test
        ## plot membrane potential trace
        figure()
        vlines(spike_time, 0, 1.5, linewidth=1)
        #plot(time, V)
        title('Leaky Integrate and Fire Neuron')
        ylabel('Membrane Potential (V)')
        xlabel('Time (sec)')
        ylim([0,2])
        show()
    

def main():
    
    neuron = spikyLIF(0.02, 0.002) 
    encoder = 1
    bias = neuron.cal_bias()
    gain = neuron.cal_gain(bias)
    print "gain = " , gain
    print "Jbias = ", bias
    
    # Demo of spike generation
    neuron2 = spikyLIF(10, 4)
    neuron2.generate_spikes(T=50, dt=0.001, J=1.4, Vth=1, V_spike=0.5, R=1)
    
    # 2a) spike output for a constant input of x=0 
    x = 0
    J = neuron.cal_current(x, gain, encoder, bias)
    neuron.generate_spikes(T=1, dt=0.001, J=J, Vth=1, V_spike=0.5, R=1)
    
    # 2b) spike output for a constant input of x=1 
    x = 1
    J = neuron.cal_current(x, gain, encoder, bias)
    neuron.generate_spikes(T=1, dt=0.001, J=J, Vth=1, V_spike=0.5, R=1)
   

    
if __name__ == "__main__": main()
```

```python
# Optimal Filter for decoding pairs of spikes
import numpy
import random
import numpy as np
import functools

class spikyLIF():
    
    def __init__(self, tau_rc=0.02, tau_ref=0.002):
        self.tau_rc = tau_rc
        self.tau_ref = tau_ref 
        
    def cal_bias(self):
        pass

    def cal_gain(self, bias): 
        pass
    
    def cal_current(self, x, gain, encoder, bias):
        J = (gain * x * encoder) + bias  
        return J
    
    def generate_spikes(self, T, dt, J, Vth=1, V_spike=0.5, R=1):
        """
        Parameters:
        ----------
        T - total simulation time (sec)    
        dt - simulation time step (sec)  
        Vth - spike threshold voltage (V)
        V_spike - spike delta voltage above Vth (V)
        R  - resistance (Ohm)
        """
        time = np.arange(0, T, dt)       # time array
        t_rest = 0                      # initial refractory time
        
        V = zeros(len(time))    # potential (V) trace over time
        S = zeros(len(time))
 
        num_spikes = 0
        # iterate over each time step
        for i, t in enumerate(time): 
            if t > t_rest:
                if numpy.isscalar(J):
                    V[i] = V[i-1] + ((J*R - V[i-1]) / self.tau_rc) * dt
                else:
                    V[i] = V[i-1] + ((J[t*1000]*R - V[i-1]) / self.tau_rc) * dt
            if V[i] >= Vth:
                num_spikes += 1
                V[i] += V_spike
                t_rest = t + self.tau_ref
                S[i] = 1.5
            if V[i] < 0:  
                V[i] = 0  
         
        
        #print "Number of Spikes = " , num_spikes
        return S
    
        
class GaussianWhiteNoise():
    
    def __init__(self, T, limit, rms):
        """
        Parameters:
        -----------
        T: the length of the signal in seconds
        limit: the maximum frequency for the signal (in Hz)
        rms: the root mean square power level of the signal.
        """
        self.T = T
        self.limit = limit
        self.rms = rms
              
    
    def create_sample(self, dt, rng):
        """
        Generate random fourier coefficients and convert them back to 
        time domain to get a signal with a limit on its max frequecy
        Parameters:
        -----------
        dt: the time step in seconds
        rng: A random number generatorn
        """
                    
        # set value for the std. deviation
        std_dev = 1
        
        # num of points(coefficients) is halved since irfft method used later returns n output points if input has n//2+1 points
        num_points = int(np.ceil(( self.T / dt) / 2.0 ))
      
        # generate the matrix containing the points
        y_values = rng.normal(0.0, std_dev, (1, num_points + 1) )
        x_values = rng.normal(0.0, std_dev, (1, num_points + 1) )
        points = x_values + 1j * y_values
        points[:, 0] = 0.0                  # set the DC component to zero 
       
             
        # Check that the max frequency is below the limit and adjust the power accordingly
        sample_freqs = np.fft.rfftfreq(num_points * 2, dt)      # compute discrete fourier transform sample frequencies  
                                                                # returns 501 frequency samples (n//2 + 1)       
        zero_points = sample_freqs > self.limit
        points[:, zero_points] = 0.0                            # set the coefficients to zero for frequencies higher than the limit
 
        a = np.array([0])
        signal = np.fft.irfft(points, axis=1)          # compute the inverse of the Discrete Fourier transform to get the real signal
        
        # compute the rms of the signal
        rms_orig = matplotlib.mlab.rms_flat(signal)
        
        # scale the signal so that it has the rms specified
        signal = signal * self.rms / rms_orig
          
        return functools.partial(self.sample_func, a, signal)


    @staticmethod
    def sample_func(a, signal):
        size =  signal.shape[1] - 1
        b = a % size
        sample = signal[:,np.array(b)[0]]
        a += 1
        return sample        


# A function to generate a randomly varying x(t) signal
def generate_signal(T, dt, rms, limit, seed):
    
    obj = GaussianWhiteNoise(T, limit, rms)
    sample = obj.create_sample(dt, rng=np.random.RandomState(seed))

    samples_array = np.arange(T/dt)             # generate evenly spaced samples
    x = np.array([sample() for i in samples_array])[:,0]
    
    X = np.fft.fftshift(np.fft.fft(x))          # compute the discrete fourier transform and 
                                                      # shift the zero frequency component to the center of the spectrum
    """# plot the time domain signal
    figure()
    plot(samples_array*dt, x)          # samples_array is multiplied by dt to limit the duration of the signal to 'T'
    xlabel('$t$')
    ylabel('$x(t)$')"""
    
    return x, X 
    


def main():
    T = 2.0         # length of signal in seconds
    dt = 0.001      # time step size
    
    # parameters for x(t)
    seed = 3     # the random number seed to use (so we can regenerate the same signal again)
    rms = 0.5    # the root mean square power level of the signa
    limit = 5    # the maximum frequency for the signal x(t) (in Hz)

    # Generate bandlimited white noise (use your own function from part 1.1)
    #x, X = syde556.generate_signal(T, dt, rms=0.5, limit=5, seed=3)
    x, X = generate_signal(T, dt, rms, limit, seed)

    Nt = len(x)                # Count the number of sample points in the time varying signal
    t = numpy.arange(Nt) * dt  # Generate evenly spaced (spaced by 1) Nt values between the interval  [0,Nt) and then multipy  
                               # by dt to change the spacing between two adjacent points to dt

    # Neuron parameters
    tau_ref = 0.002          # Absolute Refractory period of the neuron
    tau_rc = 0.02            # The RC time constant of the LIF neuron 
    x0 = 0.0                 # firing rate at x=x0 is a0
    a0 = 40.0
    x1 = 1.0                 # firing rate at x=x1 is a1
    a1 = 150.0

    # Calculate the gain, bias and the threshold of the LIF neuron
    eps = tau_rc/tau_ref
    r1 = 1.0 / (tau_ref * a0)
    r2 = 1.0 / (tau_ref * a1)
    f1 = (r1 - 1) / eps
    f2 = (r2 - 1) / eps
    alpha = (1.0/(numpy.exp(f2)-1) - 1.0/(numpy.exp(f1)-1))/(x1-x0) 
    x_threshold = x0-1/(alpha*(numpy.exp(f1)-1))              
    Jbias = 1-alpha*x_threshold;   

    # Simulate the two neurons (use your own function from part 3)
    #spikes = syde556.two_neurons(x, dt, alpha, Jbias, tau_rc, tau_ref)
    

    # Create an LIF neuron (1)
    neuron1 = spikyLIF(tau_rc, tau_ref) 
    encoder1 = 1.0
    
    # Create another LIF neuron (2)
    neuron2 = spikyLIF(tau_rc, tau_ref) 
    encoder2 = -1.0
    
    
    # generate the spikes for the first neuron
    J1 = neuron1.cal_current(x, alpha, encoder1, Jbias)
    V1 = neuron1.generate_spikes(T, dt, J=J1, Vth=1, V_spike=0.5, R=1)
    
    # generate the spikes for the second neuron
    J2 = neuron2.cal_current(x, alpha, encoder2, Jbias)
    V2 = neuron2.generate_spikes(T, dt, J=J2, Vth=1, V_spike=0.5, R=1)
    
    # create and array of spikes of both neurons
    spikes = np.array([V1, V2])
    
    freq = numpy.arange(Nt)/T - Nt/(2.0*T)   # spread the frequencies evenly from -500 to 500 (enforcing symmetry to get a real signal)
    omega = freq*2*numpy.pi                  # Calculate the corresponding angular frequencies in randians

    r = spikes[0] - spikes[1]                # Compute the resultant response of the two balanced neurons combined together
    R = numpy.fft.fftshift(numpy.fft.fft(r)) # Compute the discrete fourier transform of the response and 
                                             # then shift the zero frequency component to the centre of the spectrum  

    
    sigma_t = 0.025                          # The filter should decay towards zero. 
                                             # A spike shouldn't affect values far away from when it happens
    W2 = numpy.exp(-omega**2*sigma_t**2)     # Therefore, create a gaussian window to accomplish this
    W2 = W2 / sum(W2)                        # Scale the window appropriately

                                          # Computing H(ω)= [(X(ω)R(ω)*)∗W(ω)] / [|R(ω)|^2∗W(ω)]
    CP = X*R.conjugate()                  # Compute  X(ω)R(ω)*
    WCP = numpy.convolve(CP, W2, 'same')  # (X(ω)R(ω)*)∗ W(ω) Convolve with the gausian window
    RP = R*R.conjugate()                  # Compute |R(ω)|^2 -> Square of magnitude of R
    WRP = numpy.convolve(RP, W2, 'same')  # Compute the convolution with the gausian window |R(ω)|^2∗W(ω)
   
    XP = X*X.conjugate()                  # Compute |X(ω)|^2 -> Square of magnitude of X
    WXP = numpy.convolve(XP, W2, 'same')  # Compute the convolution with the gaussian window |X(ω)|^2∗W(ω)

    H = WCP / WRP                         # Compute the filter H(ω)= [(X(ω)R(ω)*)∗W(ω)] / [|R(ω)|^2∗W(ω)] 

    h = numpy.fft.fftshift(numpy.fft.ifft(numpy.fft.ifftshift(H))).real     # Compute the filter in time domain
                                                                            # and shift the zero frequency component to the centre
                                                                            # Only get the real part of the filter

    XHAT = H*R                            # Compute a decoded estimation of x in frequency domain ( XHAT(ω)=R(ω)H(ω) )
    

    xhat = numpy.fft.ifft(numpy.fft.ifftshift(XHAT)).real  # Compute xhat(t) by taking the inverse fourier transform of XHAT(ω)


    import pylab

    pylab.figure(0)
    pylab.subplot(1,2,1)
    pylab.plot(freq, numpy.sqrt(XP), label='$|X(\omega)|$')  # The original signal plotted in frequency domain 
    pylab.legend()                                           # magnitude of X -> [X(ω)]  versus frequency (Hz) 
    pylab.xlabel('freq (Hz)')
    pylab.ylabel('$|X(\omega)|$')
    xlim(-50, 50)
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    figure(1)
    pylab.subplot(1,2,2)
    pylab.plot(freq, numpy.sqrt(RP), label='$|R(\omega)|$')  # The response r(t)=a1(t)−a2(t) plotted in frequency domain
    pylab.legend()                                           # magnitude of R -> |R(ω)| versus frequency (Hz)
    pylab.xlabel('freq (Hz)')
    pylab.ylabel('$|R(\omega)|$')
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    pylab.figure(2)
    pylab.subplot(1,2,1)
    pylab.plot(freq, H.real)            # The Optimal filter plotted in frequency domain 
    pylab.xlabel('freq (Hz)')           # real part of H versus frequency (Hz)
    pylab.title('H (real) versus frequency')
    pylab.xlim(-50, 50)

    #figure(3)
    pylab.subplot(1,2,2)
    pylab.plot(t-T/2, h)        # The plot of the optimal filter in time domain (only the real part hence t-T/2)
    pylab.title('$h(t)$ versus time')
    pylab.xlabel('t')
    pylab.xlim(-0.5, 0.5)


    pylab.figure(4)
    pylab.plot(t, r, color='k', label='spikes', alpha=0.2)                # Plot response versus time [ r(t)=a1(t)−a2(t) ]
    pylab.plot(t, x, linewidth=2, label='input signal x(t)')              # Plot the original input signal x(t) versus time
    pylab.plot(t, xhat, label='decoded estimate xhat(t)')                     # Plot the decoded estimate xhat versus time
    pylab.title('Reconstruction of the signal using an optimal filter')
    pylab.legend(loc='best')
    pylab.xlabel('time')
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
    
    pylab.figure(5)
    pylab.subplot(1,2,1)
    pylab.plot(freq, numpy.abs(XHAT), label='$|XHAT(\omega)|$')  # The estimated signal plotted in frequency domain 
    pylab.legend()                                           # magnitude of XHAT -> [XHAT(ω)]  versus frequency (Hz) 
    pylab.xlabel('freq (Hz)')
    pylab.ylabel('$|XHAT(\omega)|$')
    xlim(-50, 50)
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
    pylab.figure(6)
    pylab.subplot(1,2,1)
    pylab.plot(freq, numpy.abs(H))           # The Optimal filter plotted in frequency domain
    pylab.xlabel('freq (Hz)')           # magnitude of H -> |H(ω)| versus frequency (Hz)
    pylab.title('$|H(\omega)|$ versus frequency')
    pylab.xlim(-50, 50)
    
    pylab.show()

if __name__ == "__main__": main()
```
