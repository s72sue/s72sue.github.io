```python
%pylab inline
import random
import numpy as np
import functools
import matplotlib.pyplot as plt


class GaussianWhiteNoise():
    
    def __init__(self, T, limit, rms):
        """
        Parameters:
        -----------
        T: the length of the signal in seconds
        limit: the maximum frequency for the signal (in Hz)
        rms: the root mean square power level of the signal.
        """
        self.T = T
        self.limit = limit
        self.rms = rms
              
    
    def create_sample(self, dt, rng):
        """
        Generate random fourier coefficients and convert them back to 
        time domain to get a signal with a limit on its max frequecy
        Parameters:
        -----------
        dt: the time step in seconds
        rng: A random number generatorn
        """
                    
        # set value for the std. deviation
        std_dev = 1
        
        # num of points(coefficients) is halved since irfft method used later returns n output points if input has n//2+1 points
        num_points = int(np.ceil(( self.T / dt) / 2.0 ))
      
        # generate the matrix containing the points
        y_values = rng.normal(0.0, std_dev, (1, num_points + 1) )
        x_values = rng.normal(0.0, std_dev, (1, num_points + 1) )
        points = x_values + 1j * y_values
        points[:, 0] = 0.0                  # set the DC component to zero 
       
             
        # Check that the max frequency is below the limit and adjust the power accordingly
        sample_freqs = np.fft.rfftfreq(num_points * 2, dt)      # compute discrete fourier transform sample frequencies  
                                                                # returns 501 frequency samples (n//2 + 1)       
        zero_points = sample_freqs > self.limit
        points[:, zero_points] = 0.0                            # set the coefficients to zero for frequencies higher than the limit
 
        a = np.array([0])
        signal = np.fft.irfft(points, axis=1)          # compute the inverse of the Discrete Fourier transform to get the real signal
        
        # compute the rms of the signal
        rms_orig = matplotlib.mlab.rms_flat(signal)
        
        # scale the signal so that it has the rms specified
        signal = signal * self.rms / rms_orig
          
        return functools.partial(self.sample_func, a, signal)


    @staticmethod
    def sample_func(a, signal):
        size =  signal.shape[1] - 1
        b = a % size
        sample = signal[:,np.array(b)[0]]
        a += 1
        return sample

    
# A function to generate a randomly varying x(t) and X(w) signal
def generate_signal(T, dt, rms, limit, seed):
    
    obj = GaussianWhiteNoise(T, limit, rms)
    sample = obj.create_sample(dt, rng=np.random.RandomState(seed))

    samples_array = np.arange(T/dt)             # generate evenly spaced samples
    x = np.array([sample() for i in samples_array])[:,0]
    
    X = np.fft.fftshift(np.fft.fft(x))          # compute the discrete fourier transform and 
                                                      # shift the zero frequency component to the center of the spectrum
    # plot the time domain signal
    figure()
    plot(samples_array*dt, x, 'b')          # samples_array is multiplied by dt to limit the duration of the signal to 'T'
    xlabel('$t$')
    ylabel('$x(t)$')
    
    return x, X   


# Leaky Integrate and Fire Neuron Model

class LeakyIntegNeuron ():
     
    def __init__(self, tau_ref=0.002, tau_rc=0.02):
        self.tau_rc = tau_rc
        self.tau_ref = tau_ref
    
    def rates (self, x, gain, bias, S, encoder):
        J = (gain * x * encoder) + bias       # compute the current using the gain and the bias
        a = np.linspace(0,0,S)
        try:  
            i = 0 
            for element in J:
                if element > 1:
                    a[i] = 1.0 / (self.tau_ref - self.tau_rc * math.log( 1 - (1/element) ) )
                i = i+1    
        finally:
            np.seterr(divide='ignore', invalid='ignore')
        return a

    def cal_bias(self, intercept, gain):
        bias = 1 - (gain*intercept)  
        return bias

    def cal_gain(self, intercept, max_firing_rate):
        try:
            x = 1.0 / (1 - np.exp((self.tau_ref - (1.0 / max_firing_rate)) / self.tau_rc))
            gain = ( (1-x) / (intercept - 2.0) )       # (intercept - 2) since the maxfiring rate occurs at x = 2
        finally:
            np.seterr(divide='ignore')
        return gain
    
    def cal_current(self, x, gain, encoder, bias):
        J = (gain * x * encoder) + bias  
        return J
    
    def generate_spikes(self, T, dt, J, Vth=1, V_spike=0.5, R=1):
        """
        Parameters:
        ----------
        T - total simulation time (sec)    
        dt - simulation time step (sec)  
        Vth - spike threshold voltage (V)
        V_spike - spike delta voltage above Vth (V)
        R  - resistance (Ohm)
        """
        time = np.arange(0, T, dt)       # time array
        t_rest = 0                      # initial refractory time
        
        V = zeros(len(time))    # potential (V) trace over time
        S = zeros(len(time))    # spikes
 
        num_spikes = 0
        # iterate over each time step
        for i, t in enumerate(time): 
            if t > t_rest:
                if numpy.isscalar(J):
                    V[i] = V[i-1] + ((J*R - V[i-1]) / self.tau_rc) * dt
                else:
                    V[i] = V[i-1] + ((J[t*(T/dt)]*R - V[i-1]) / self.tau_rc) * dt
            if V[i] >= Vth:
                num_spikes += 1
                V[i] += V_spike
                t_rest = t + self.tau_ref
                S[i] = 1.5
            if V[i] < 0:  
                V[i] = 0  
         
                
        #print "Number of Spikes = " , num_spikes
        return S
 

def cal_decoders(activity_matrix, x, S, std_deviation):
    gamma = (( np.dot(activity_matrix, activity_matrix.transpose()) ) / S )            # Y = AA^T/S
   
    # if standard deviation is not zero, then consider the effect of noise
    if std_deviation:  
        gamma = gamma + (std_deviation**2)
           
    upsilon = ( np.dot(activity_matrix, x) ) / S         # v = A.x/S
    try:
        decoders = np.dot(np.linalg.pinv(gamma), upsilon)        # compute the decoders (using psedo inverse to deal with singular matrices)
    finally:
        np.seterr(divide='ignore', invalid='ignore')  
    return decoders 


# computes the xhat vector 
def compute_xhat(x, decoders, activity_matrix, S):
    xhat = np.linspace(0,0,S)
    i = 0
    for xvalues in x:
        xhat[i] = np.dot(activity_matrix[:,i], decoders)
        i = i+1  
    return xhat


# generate error plots and compute RMSE value
def plot_error(x, xhat, append):
    
    plt.figure()        # create a new plot/figure to overlay the plot y=x   
    plt.plot(x,x)
    plt.plot(x,xhat)
    plt.xlabel('$x$')
    plt.ylabel('$\hat{x}$')
    #plt.ylim(-1, 1)
    #plt.xlim(-1, 1)
    plt.title(append + 'y = x : Overlaid')
    
    plt.figure()        # create a new plot/figure for plotting the error   
    plt.plot(x,x-xhat)
    plt.xlabel('$x$')
    plt.ylabel('$\hat{x}-x$')
    #plt.ylim(-1, 1)
    #plt.xlim(-1, 1)
    plt.title(append + 'Error Plot')
    
    rmse = np.sqrt(np.average((x-xhat)**2))            # calculate the root mean square error
    return rmse
     

def compute_filter(dt, tau, n, T, colour):
    
    t = np.arange(T/dt)*dt
    t = t - T/2
    h = (t**n) * np.exp(-t/tau)  
    h[np.where(t<0)]=0
    h = h/norm(h,1)

    #figure()
    plot(t, h, colour)
    xlabel('t')
    ylabel('h(t)')
    xlim(-0.05, 0.08)
    
    return h


def main():
    
    num_neurons = 20 
    intercepts = np.random.uniform(-2,2,num_neurons)       # intercepts between -2 and 2  ( a+(b-a)*rand )
    max_firing_rates = np.random.uniform(100,200,num_neurons)     # max firing rates of neurons should be between 100 to 200 Hz
    encoders = [random.choice([-1,1]) for count in xrange(num_neurons)] 
    neurons = [LeakyIntegNeuron() for count in xrange(num_neurons)]       # create 16 rectified linear neurons
      
    S = 80      # number of samples so that dx = 0.05
    x = np.linspace(-2,2,S)
    activity_matrix = np.zeros((num_neurons,S))       #creates a matrix with dimentions 16x40, each element initilized to zero
    plt.figure()                             # creates a new figure for plotting firing rates
    i = 0
    for neuron in neurons:
        gain = neuron.cal_gain(intercepts[i], max_firing_rates[i])
        bias = neuron.cal_bias(intercepts[i], gain)
        firing_rates = neuron.rates(x, gain, bias, S, encoders[i])     # firing rates of a neuron over the entire range (range of x values)
        activity_matrix[i,:] = firing_rates        # populate the firing rates of this neuron at every value of x 
        i = i + 1    
    
    
# ------------------------------------------------------------------------------------------------------------------------------------

    # select a neuron having firing rate between 20 HZ and 50 Hz at x = 0
    sel_activity = NAN
    sel_neuron_index = NAN
    i = 0
    for neuron in neurons: 
        if ( activity_matrix[i,40] > 20 and activity_matrix[i,40] < 50):
            sel_activity = activity_matrix[i,40]
            sel_neuron_index = i
        i = i + 1
     
    print activity_matrix[:,40]
    print "\nselected rate: ", sel_activity, "\nselected neuron index: ",  sel_neuron_index    
        
    sel_gain = neuron.cal_gain(intercepts[sel_neuron_index], max_firing_rates[sel_neuron_index])
    sel_bias = neuron.cal_bias(intercepts[sel_neuron_index], sel_gain)
    
    # construct two neurons with same gain and bias as the neuron selected
    neuron1 = LeakyIntegNeuron()  
    firing_rate1 = neuron1.rates(x, sel_gain, sel_bias, S, 1)
    neuron2 = LeakyIntegNeuron()
    firing_rate2 = neuron2.rates(x, sel_gain, sel_bias, S, -1)
    figure()
    plt.plot(x, firing_rate1, 'r')
    plt.plot(x, firing_rate2, 'g')
    #plt.plot(x, activity_matrix[sel_neuron_index,:], 'b')
    plt.xlabel('$x$')
    plt.ylabel('$a$ (Hz)')
    
    
 # -------------------------------------------------------------------------------------------------------------------------------------   
    
    # Generate a time varying signal
    T = 1.0
    dt = 0.001
    seed = 3
    rms = 1
    limit = 5
    x, X = generate_signal(T, dt, rms, limit, seed) 
    
    Nt = len(x)                # Count the number of sample points in the time varying signal
    t = numpy.arange(Nt) * dt  # Generate evenly spaced (spaced by 1) Nt values between the interval  [0,Nt) and then multipy  
                               # by dt to change the spacing between two adjacent points to dt
   
    # feed the signal into the two neurons and generate spikes
    J1 = neuron1.cal_current(x, sel_gain, 1, sel_bias)
    spikes1 = neuron1.generate_spikes(T, dt, J=J1, Vth=1, V_spike=0.5, R=1)
    plot(t,spikes1, 'r')
    
    J2 = neuron1.cal_current(x, sel_gain, -1, sel_bias)
    spikes2 = neuron1.generate_spikes(T, dt, J=J2, Vth=1, V_spike=0.5, R=1)
    plot(t,spikes2-1.5, 'g') 
    
    # Part a) -  plot h(t)
    tau = 0.005
    n = 0.0
    figure()
    h = compute_filter(dt, tau, n, T, 'b')
    
    # decode the spikes back to xhat(t)
    fspikes1 = np.convolve(spikes1, h, mode='same')
    fspikes2 = np.convolve(spikes2, h, mode='same')

    A = np.array([fspikes1, fspikes2])
    decoders = cal_decoders(A, x, Nt, 0)
    xhat = np.dot(A.T, decoders)
    
    #  Part b) - Plot x(t), spikes and decoded xhat value
    figure()
    plot(t,spikes1, 'r')
    plot(t,spikes2-1.5, 'g')
    plot(t,xhat, 'cyan')
    plot(t,x,'b')
    xlabel('$t$')
    ylabel('$x(t)$')
    
    # Part C - Compute the root mean square of the decoding
    rmse = np.sqrt(np.average((x-xhat)**2)) 
    print "RMSE of the original signal:", rmse
    
if __name__ == "__main__": main()   
```
