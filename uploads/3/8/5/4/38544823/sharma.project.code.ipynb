# Wason Selection Task

### Create the core Wason Model

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from nengo.dists import Uniform
from nengo.spa import Vocabulary
from nengo.spa import SemanticPointer
from nengo import spa
import random
import nengo
import nengo.spa


#-----------------------------------------------------------------------------------------------------------------
#****************************EXPERIMENTAL MODES**********************************

# exp_mode 1 is for training the system in different contexts
# exp_mode 2 is for syntactic generalization within a given context
exp_mode = 2    

# sub_exp_mode is a sub mode of exp_mode 2 (i.e., the generalization experiment)
# sub_exp_mode 2 - two examples are presented for generalization
# sub_exp_mode 3 - three examples are presented for generalization
# sub_exp_mode 4 - four examples are presented for generalization
sub_exp_mode = 2        


#-----------------------------------------------------------------------------------------------------------------
#************************************PARAMETERS***********************************

D=32                    # number of dimensions per ensemble
N=10                    # number of neurons per dimension for the representaion populations
N_mem = 40              # number of neurons per dimension for the memory populations
N_conv=100              # number of neurons per dimension for the convolution populations
N_other= N*D            # total number of neurons for representation populations
learning_rate=5e-5      # PES learning rate of the model



#-----------------------------------------------------------------------------------------------------------------
#***********************************Neuron Modes************************************

mode_switch = True                   # used to set the mode for the memory populations, set to 'True' for direct mode
neuron_mode = nengo.Direct()         # used to set the mode for the representation populations, set to 'nengo.Direct()' for direct mode
lif_mode = nengo.LIF()               # used to set the mode for the learning populations, this should always be set to LIF mode
if neuron_mode == nengo.Direct():    # Number of neurons used in direct mode do not matter, so set them to 1
    N_other = 1
    N_conv = 1
    N_mem = 1
    N = 1
    

#-----------------------------------------------------------------------------------------------------------------
#***********************************Create the Core Wason Model************************************
    
model = spa.SPA('Wason')
with model:  
    context = nengo.Ensemble(label='context', n_neurons=N, dimensions=1, neuron_type=lif_mode)
    rule = nengo.Ensemble(label='rule', n_neurons=N_other, dimensions=D, neuron_type=neuron_mode)
    transform = nengo.Ensemble(label='transform', n_neurons=N_other, dimensions=D, neuron_type=lif_mode)
    answer = nengo.Ensemble(label='answer', n_neurons=N_other, dimensions=D, neuron_type=neuron_mode) 
    
    correct_transform = nengo.Ensemble(label='correct transform', n_neurons=N_other, dimensions=D, neuron_type=lif_mode)
    correct_answer = nengo.Ensemble(label='correct answer', n_neurons=N_other, dimensions=D, neuron_type=neuron_mode)
    
    # computing R' convolved with A* (correct answer)
    cconv2 = nengo.networks.CircularConvolution(n_neurons=D*N_conv, dimensions=D, invert_b=True)
    nengo.Connection(correct_answer, cconv2.A)
    nengo.Connection(rule, cconv2.B)
    nengo.Connection(cconv2.output, correct_transform)
    
    # computing  Answer = T convolved with R
    cconv = nengo.networks.CircularConvolution(n_neurons=D*N_conv, dimensions=D)
    nengo.Connection(rule, cconv.A)
    nengo.Connection(transform, cconv.B)
    nengo.Connection(cconv.output, answer) 
       
    #conn = nengo.Connection(context, transform, function=lambda x: np.random.random(D))    #PES
    conn = nengo.Connection(context, transform, function=lambda x: np.random.random(D), solver=nengo.solvers.LstsqL2(weights=True))  #HPES
    error = nengo.Ensemble(N_other, dimensions=D)
    
    # Error = pre - post (correct transform - transform)
    nengo.Connection(transform, error, transform=-1)
    nengo.Connection(correct_transform, error)
    
    # Modulatory connections don't impart current
    error_conn = nengo.Connection(error, transform, modulatory=True)
    
    # Add the learning rule to the connection
    #conn.learning_rule_type = nengo.PES(error_conn, learning_rate=learning_rate)   #PES
    conn.learning_rule_type = [nengo.PES(error_conn, learning_rate=learning_rate), nengo.BCM(learning_rate=1e-10)]   #HPES
```

### Provide Input to the model

```python
# Change the seed of the rng to change the vocabulary
rng = np.random.RandomState(3)

# create a vocabulary
vocab = Vocabulary(dimensions=D, rng=rng, max_similarity=0.05)  

#-----------------------------------------------------------------------------------------------------------------
#***************************************Experimental Mode 1 - Learning********************************************
    
if exp_mode==1:
      
    # If a card has a vowel on one side, it must have an even number on the other side.
    vowel = vocab.parse('TYPE*VOWEL+CATEGORY*ALPHABET')
    RULE1 = vocab.parse('ANTE') * vowel + vocab.parse('REL*IMPL+CONS*EVEN')
    
    # If you can vote, you must be over 18 years of age.
    vote = vocab.parse('ACT*VOTE+PREREQUISITE*AGE')
    RULE2 = vocab.parse('ANTE') * vote + vocab.parse('REL*IMPL+CONS*OVER18')

    test_rule1 = RULE1
    test_rule2 = RULE2
    ANSWER1=vocab.parse('VOWEL+EVEN')
    ANSWER2=vocab.parse('VOTE+NOT*OVER18')
    vocab.add('NOT_OVER18',vocab.parse('NOT*OVER18'))
    ANSWER1.normalize()
    ANSWER2.normalize()
    RULE1.normalize()
    RULE2.normalize()
    

    with model:
        if test_rule1.compare(RULE1) > test_rule1.compare(RULE2):
            cont1 = -1 * test_rule1.compare(RULE1)
        else:
            cont1 = test_rule1.compare(RULE2)
        
        if test_rule2.compare(RULE1) > test_rule2.compare(RULE2):
            cont2 = -1 * test_rule2.compare(RULE1)
        else:
            cont2 = test_rule2.compare(RULE2)
            
        determined_context1 = nengo.Node(output=lambda t: cont1 if (t>3.0 and t<=4.5) else 0.0, size_out=1)
        determined_context2 = nengo.Node(output=lambda t: cont2 if (t>4.5 and t<=6) else 0.0, size_out=1)
        abstract_context = nengo.Node(output=lambda t: -1 if t <= 1.5 else 0.0, size_out=1)
        familiar_context = nengo.Node(output=lambda t: 1 if (t>1.5 and t<=3.0) else 0.0, size_out=1)
        rule1 = nengo.Node(output=lambda t: RULE1.v if t <= 1.5 or (t>3.0 and t<=4.5) else 0.0, size_out=D)
        rule2 = nengo.Node(output=lambda t: RULE2.v if (t>1.5 and t<=3) or (t>4.5 and t<=6) else 0.0, size_out=D)
        answer1 = nengo.Node(output=lambda t: ANSWER1.v if t <= 1.5 or (t>3.0 and t<=4.5) else 0.0, size_out=D)
        answer2 = nengo.Node(output=lambda t: ANSWER2.v if (t>1.5 and t<=3) or (t>4.5 and t<=6) else 0.0, size_out=D)
        
        nengo.Connection(familiar_context, context)
        nengo.Connection(abstract_context, context)
        nengo.Connection(determined_context1, context)
        nengo.Connection(determined_context2, context)
        nengo.Connection(rule1, rule)
        nengo.Connection(rule2, rule)
        nengo.Connection(answer1, correct_answer)
        nengo.Connection(answer2, correct_answer)
        

#------------------------------------------------------------------------------------------------------------------
#**********************************Experimental Mode 2 - Syntactic Generalization**********************************        
    
elif exp_mode==2:
    
    vocab.add('NOT_OVER16',vocab.parse('NOT*OVER16'))
    vocab.add('NOT_OVER18',vocab.parse('NOT*OVER18'))
    vocab.add('NOT_OVER21',vocab.parse('NOT*OVER21'))
    
    # If you drink, you must be over 21 years of age.
    drink = vocab.parse('ACT*DRINK+PREREQUISITE*AGE')
    RULE1 = vocab.parse('ANTE') * drink + vocab.parse('REL*IMPL+CONS*OVER21') 

    # If you drive, you must be over 16 years of age.
    drive = vocab.parse('ACT*DRIVE+PREREQUISITE*AGE')
    RULE3 = vocab.parse('ANTE') * drive + vocab.parse('REL*IMPL+CONS*OVER16')
    
    # If you can vote, you must be over 18 years of age.
    vote = vocab.parse('ACT*VOTE+PREREQUISITE*AGE')
    RULE2 = vocab.parse('ANTE') * vote + vocab.parse('REL*IMPL+CONS*OVER18')
    
    ANSWER2=vocab.parse('VOTE+NOT*OVER18')
    ANSWER3=vocab.parse('DRIVE+NOT*OVER16')
    ANSWER1=vocab.parse('DRINK+NOT*OVER21')  
    ANSWER1.normalize()
    ANSWER2.normalize()
    ANSWER3.normalize()
    RULE1.normalize()
    RULE2.normalize()
    RULE3.normalize()
    
    scaling_factor = 0.5
    
    if sub_exp_mode == 3:
        n_rules = 3.0 
        
        # If are going for skydiving, you must be over 16 years of age
        skydiving = vocab.parse('ACT*SKYDIVING+PREREQUISITE*AGE')
        RULE4 = vocab.parse('ANTE') * skydiving + vocab.parse('REL*IMPL+CONS*OVER16')
        
        cont = ( (RULE3.compare(RULE1) + RULE3.compare(RULE2) + RULE3.compare(RULE4)) * (1/3.0) ) * (n_rules*scaling_factor)
        ANSWER4=vocab.parse('SKYDIVING+NOT*OVER16')
        ANSWER4.normalize()
        RULE4.normalize()
        
        
    elif sub_exp_mode == 4:
        n_rules = 4.0
        
        # If are going for skydiving, you must be over 16 years of age
        skydiving = vocab.parse('ACT*SKYDIVING+PREREQUISITE*AGE')
        RULE4 = vocab.parse('ANTE') * skydiving + vocab.parse('REL*IMPL+CONS*OVER16') 
        
        # If you are applying for Canadian Citizenship, you must be over 18 years of age.
        citizenship = vocab.parse('ACT*CITIZENSHIP+PREREQUISITE*AGE')
        RULE5 = vocab.parse('ANTE') * citizenship + vocab.parse('REL*IMPL+CONS*OVER18')
        
        cont = ( (RULE3.compare(RULE1) + RULE3.compare(RULE2) + RULE3.compare(RULE4) + RULE3.compare(RULE5)) * (1/4.0) ) * (n_rules*scaling_factor)
        ANSWER4=vocab.parse('SKYDIVING+NOT*OVER16')
        ANSWER5=vocab.parse('CITIZENSHIP+NOT*OVER18')
        ANSWER4.normalize()
        ANSWER5.normalize()
        RULE4.normalize()
        RULE5.normalize()
        
        
    else:
        n_rules = 2.0
        cont = ( (RULE3.compare(RULE1) + RULE3.compare(RULE2)) * (1/2.0) ) * (n_rules*scaling_factor)
        

    with model:
        
        if sub_exp_mode == 3:
            rule4 = nengo.Node(output=lambda t: RULE4.v if (t>3 and t<=4.5) else 0.0, size_out=D) 
            answer4 = nengo.Node(output=lambda t: ANSWER4.v if (t>3 and t<=4.5) else 0.0, size_out=D)     
            nengo.Connection(rule4, rule)
            nengo.Connection(answer4, correct_answer)
                 
        elif sub_exp_mode == 4:
            rule4 = nengo.Node(output=lambda t: RULE4.v if (t>3 and t<=4.5) else 0.0, size_out=D) 
            answer4 = nengo.Node(output=lambda t: ANSWER4.v if (t>3 and t<=4.5) else 0.0, size_out=D)  
            rule5 = nengo.Node(output=lambda t: RULE5.v if (t>4.5 and t<=6) else 0.0, size_out=D) 
            answer5 = nengo.Node(output=lambda t: ANSWER5.v if (t>4.5 and t<=6) else 0.0, size_out=D)        
            nengo.Connection(rule4, rule)
            nengo.Connection(rule5, rule)
            nengo.Connection(answer4, correct_answer)
            nengo.Connection(answer5, correct_answer)
            
        
        familiar_context = nengo.Node(output=lambda t: 1 if t<6 else 0.0, size_out=1)
        determined_context = nengo.Node(output=lambda t: cont if (t>6 and t<=7.5) else 0.0, size_out=1)
        rule1 = nengo.Node(output=lambda t: RULE1.v if t <= 1.5 else 0.0, size_out=D)
        rule2 = nengo.Node(output=lambda t: RULE2.v if (t>1.5 and t<=3) else 0.0, size_out=D) 
        rule3 = nengo.Node(output=lambda t: RULE3.v if (t>6 and t<=7.5) else 0.0, size_out=D)  
        answer1 = nengo.Node(output=lambda t: ANSWER1.v if t <= 1.5 else 0.0, size_out=D)
        answer2 = nengo.Node(output=lambda t: ANSWER2.v if (t>1.5 and t<=3) else 0.0, size_out=D)
        answer3 = nengo.Node(output=lambda t: ANSWER3.v if (t>6 and t<=7.5) else 0.0, size_out=D) 
        
        nengo.Connection(familiar_context, context)
        nengo.Connection(determined_context, context)
        nengo.Connection(rule1, rule)
        nengo.Connection(rule2, rule)
        nengo.Connection(rule3, rule)
        nengo.Connection(answer1, correct_answer)
        nengo.Connection(answer2, correct_answer)
        nengo.Connection(answer3, correct_answer)
```

### Create the Basal Ganglia and memory components

```python
sub_dim = 16
with model:
    model.answer1_memory = spa.Memory(dimensions=D,synapse=0.4, neurons_per_dimension=N_mem, vocab=vocab, subdimensions=sub_dim, direct = mode_switch)
    model.answer2_memory = spa.Memory(dimensions=D,synapse=0.4, neurons_per_dimension=N_mem, vocab=vocab, subdimensions=sub_dim, direct = mode_switch)
    model.answer_memory = spa.Buffer(dimensions=D, neurons_per_dimension=N, vocab=vocab, subdimensions=sub_dim, direct = mode_switch)
    model.out = spa.Buffer(dimensions=D, neurons_per_dimension=N, vocab=vocab, subdimensions=sub_dim, direct = mode_switch)
    model.inhibit = spa.Buffer(dimensions=1)
    model.cortex = spa.Buffer(dimensions=D,neurons_per_dimension=N, subdimensions=sub_dim, direct = mode_switch)
    model.sensory_input = spa.Buffer(dimensions=D,neurons_per_dimension=N, subdimensions=sub_dim, direct = mode_switch)
    
    # possible actions to be selected based on the utility
    actions = spa.Actions(
        'dot(cortex, A) --> inhibit = 0, out = 0',
        'dot(cortex, B) --> inhibit = 2, out = 0',
        'dot(cortex, C) --> inhibit = 2, answer1_memory = answer_memory, out = 0',
        'dot(cortex, D) --> inhibit = 2, answer2_memory = answer_memory, out = 0',
        'dot(cortex, E) --> inhibit = 2, out = answer1_memory',
        'dot(cortex, F) --> inhibit = 2, out = answer2_memory',
    )
    
    nengo.Connection(model.sensory_input.state.output, model.cortex.state.input, synapse=0.1)
    nengo.Connection(answer, model.answer_memory.state.input, synapse=0.1)
    
    model.bg = spa.BasalGanglia(actions=actions)
    model.thal = spa.Thalamus(model.bg)
    nengo.Connection(model.inhibit.state.output, error.neurons, transform=[[-1]] * error.n_neurons)
```

### Provide sensory input which determines the state of the cortex

```python
if exp_mode==1:
    def state(t):
        if t <= 3:
            return 'A'          # start learning from the rules presented                         
        elif (3 < t<=4.5):      # stop learning and determine the answer to the abstract rule presented
            return 'C'
        elif (4.5 < t<=6):      # stop learning and determine the answer to the familiar rule presented
            return 'D'
        elif (7<t<=8.5):
            return 'E'          # tell the answer for the abstract rule 
        elif (9<t<=10.5):
            return 'F'          # tell the answer for the familiar rule 
        else:
            return 'B'          # resting state, no learning and no output
        
     
if exp_mode==2:  
    def state(t):
        if t <= 6:
            return 'A'          # start learning from the rules presented  
        elif (6<t<=7.5):        # stop learning and determine the answer to the test rule presented 
            return 'D'
        elif (8.5<t<=10):
            return 'F'          # tell the answer for the test rule presented 
        else:
            return 'B'          # resting state, no learning and no output    

with model:
    model.input = spa.Input(sensory_input=state)     # provide sensory input which determines cortex state 
```

### Add probes to collect data

```python
with model:
    # Core wason model
    answer_probe = nengo.Probe(answer, synapse=0.1)
    rule_probe = nengo.Probe(rule, synapse=0.1)
    transform_probe = nengo.Probe(transform, synapse=0.1)
    
    correct_answer_probe = nengo.Probe(correct_answer, synapse=0.1)
    correct_transform_probe = nengo.Probe(correct_transform, synapse=0.1)
    
    context_probe = nengo.Probe(context, synapse=0.1)
    
    # basal ganglia performance
    cortex_probe = nengo.Probe(model.cortex.state.output, synapse=0.1)
    actions_probe = nengo.Probe(model.thal.actions.output, synapse=0.1)
    utility_probe = nengo.Probe(model.bg.input, synapse=0.1)
    
    error_probe = nengo.Probe(error, synapse=0.1)
    
    # memory components and output
    output = nengo.Probe(model.out.state.output, synapse=0.1)
    ans_memory = nengo.Probe(model.answer_memory.state.output, synapse=0.1)
    ans1_memory = nengo.Probe(model.answer1_memory.state.output, synapse=0.1)
    ans2_memory = nengo.Probe(model.answer2_memory.state.output, synapse=0.1)
```

```python
sim = nengo.Simulator(model)
sim.run(11.0)
```

### Plot the results

```python
plt.figure(figsize=(8, 4))
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[answer_probe], vocab))
plt.xlabel("t [s]")
plt.title("Computed Answer")
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)

plt.figure()
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[correct_answer_probe], vocab))
plt.title("Correct Answer (for the purpose of comparing)")
plt.xlabel("t [s]")
plt.ylabel("dot product")
plt.legend(vocab.keys, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
```

```python
plt.figure(figsize=(18, 3))
plt.subplot(1, 3, 1)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[ans1_memory], vocab))
plt.title("Answer in Memory (abstract context)")
plt.xlabel("t [s]")

plt.subplot(1, 3, 2)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[ans2_memory], vocab))
plt.title("Answer in Memory (familiar context)")
plt.xlabel("t [s]")

plt.subplot(1, 3, 3)
plt.plot(sim.trange(), nengo.spa.similarity(sim.data[output], vocab))
plt.xlabel("t [s]")
plt.title("Answer Output")
```

```python
fig = figure(figsize=(10,8))
p1 = fig.add_subplot(3,1,1)

p1.plot(sim.trange(), model.similarity(sim.data, cortex_probe))
p1.legend(model.get_output_vocab('cortex').keys, fontsize='x-small')
p1.set_ylabel('State')
p1.set_ylim(-0.5,1.5)

p3 = fig.add_subplot(3,1,2)
p3.plot(sim.trange(), sim.data[utility_probe])
p3_legend_txt = [a.condition for a in model.bg.actions.actions]
p3.legend(p3_legend_txt, fontsize='x-small')
p3.set_ylabel('Utility')
p3.set_ylim(-0.5,1.5)
p3.legend(p3_legend_txt, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)

p2 = fig.add_subplot(3,1,3)
p2.plot(sim.trange(), sim.data[actions_probe])
p2_legend_txt = [a.effect for a in model.bg.actions.actions]
p2.legend(p2_legend_txt, fontsize='x-small')
p2.set_ylabel('Action')
p2.legend(p2_legend_txt, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
p2.set_ylim(-0.5,1.5)


fig.subplots_adjust(hspace=0.2)
```
